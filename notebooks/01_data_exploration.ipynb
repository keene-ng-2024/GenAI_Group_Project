# 01 – Data Exploration
Inspect the raw review files and understand the dataset structure.

```python
import json
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
REVIEWS_PATH = Path('../data/processed/reviews_parsed.json')

with open(REVIEWS_PATH) as f:
    papers = json.load(f)

print(f'Total papers: {len(papers)}')
```

```python
rows = []
for paper_id, paper in papers.items():
    reviews = paper.get('reviews', [])
    rows.append({
        'paper_id': paper_id,
        'title': paper.get('title', '')[:60],
        'n_reviews': len(reviews),
        'mean_rating': sum(r['rating'] for r in reviews if r.get('rating')) / max(len(reviews), 1),
        'total_review_chars': sum(len(r['text']) for r in reviews),
    })

df = pd.DataFrame(rows)
df.head(10)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

df['n_reviews'].hist(ax=axes[0], bins=10, edgecolor='white')
axes[0].set_title('Reviews per paper')
axes[0].set_xlabel('# reviews')

df['total_review_chars'].hist(ax=axes[1], bins=15, edgecolor='white', color='coral')
axes[1].set_title('Total review length (chars)')
axes[1].set_xlabel('Characters')

plt.tight_layout()
plt.show()
```
